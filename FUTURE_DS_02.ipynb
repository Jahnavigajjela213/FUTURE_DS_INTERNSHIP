```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

pd.set_option("display.max_columns", 50)
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

campaign_data = {
    "Campaign": ["FB_Summer", "FB_Winter", "IG_Fashion", "IG_Tech", "FB_Festival"],
    "Impressions": [10000, 8000, 12000, 15000, 11000],
    "Clicks": [800, 600, 1000, 900, 950],
    "Conversions": [120, 90, 150, 130, 140],
    "Spend": [500, 400, 700, 650, 600]
}

df_campaign = pd.DataFrame(campaign_data)

df_campaign["CTR (%)"] = (df_campaign["Clicks"] / df_campaign["Impressions"]) * 100
df_campaign["CPC ($)"] = df_campaign["Spend"] / df_campaign["Clicks"]
df_campaign["ROI"] = (df_campaign["Conversions"] * 10 - df_campaign["Spend"]) / df_campaign["Spend"] * 100  # Assuming $10 per conversion

print("Campaign Data with KPIs:")
print(df_campaign)

plt.figure(figsize=(8,5))
sns.barplot(x="Campaign", y="CTR (%)", data=df_campaign)
plt.title("CTR by Campaign")
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(x="Campaign", y="ROI", data=df_campaign)
plt.title("ROI by Campaign")
plt.show()
```
