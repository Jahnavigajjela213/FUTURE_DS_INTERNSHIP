```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

pd.set_option("display.max_columns", 50)
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

from textblob import TextBlob

feedback = [
    "The event was amazing and well-organized!",
    "I didn’t like the food arrangements.",
    "Great speakers, but the sessions were too long.",
    "Excellent networking opportunities.",
    "The venue was too crowded and uncomfortable."
]

df_feedback = pd.DataFrame(feedback, columns=["Feedback"])


df_feedback["Polarity"] = df_feedback["Feedback"].apply(lambda x: TextBlob(x).sentiment.polarity)
df_feedback["Sentiment"] = df_feedback["Polarity"].apply(
    lambda x: "Positive" if x > 0 else ("Negative" if x < 0 else "Neutral")
)

print(df_feedback)

plt.figure(figsize=(6,4))
sns.countplot(x="Sentiment", data=df_feedback, palette="Set2")
plt.title("Feedback Sentiment Distribution")
plt.show()

avg_polarity = df_feedback["Polarity"].mean()
print("Average Polarity Score:", avg_polarity)
```
